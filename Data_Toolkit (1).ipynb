#Theory Questions:


1. What is NumPy, and why is it widely used in Python?
  - NumPy is a Python library used for numerical computing. It provides fast and efficient arrays, mathematical functions, and tools for scientific calculations.

2. How does broadcasting work in NumPy?
  - Broadcasting lets NumPy do math on arrays of different shapes by “stretching” the smaller one to match the bigger one without real copying.

3. What is a Pandas DataFrame?
  - A DataFrame is a 2-dimensional table in Pandas that stores data in rows and columns, similar to an Excel sheet or SQL table.

4. Explain the use of the groupby() method in Pandas.
  - groupby() is used to split data into groups, apply calculations (like sum or mean), and combine the results.

5. Why is Seaborn preferred for statistical visualizations.
  - Seaborn provides beautiful, ready-made statistical plots and works seamlessly with Pandas DataFrames.

6. What are the differences between NumPy arrays and Python lists?
  - Data Type:
    - NumPy arrays store elements of the same data type, whereas Python lists can store different data types.

  - Performance:
    - NumPy arrays are much faster for numerical operations because they are implemented in C, while Python lists are slower.

  - Memory Usage:
    - NumPy arrays use less memory as they store data in a compact, contiguous block. Lists require more memory due to references to objects.

  - Operations:
    - NumPy supports vectorized operations, allowing mathematical operations on entire arrays without loops. Lists require explicit looping.

  - Functionality:
    - NumPy provides many built-in mathematical and scientific functions, while Python lists offer limited numerical functionality.

7. What is a heatmap, and when should it be used?
  - A heatmap shows numbers in a table using colors instead of text. It is used to see patterns in things like correlation or confusion matrices.

8. What does the term “vectorized operation” mean in NumPy?
  - Vectorized operations allow performing calculations on entire arrays at once without using loops.

9. How does Matplotlib differ from Plotly?
  - Interactivity:
    - Matplotlib creates static plots, while Plotly generates interactive plots with features like zooming, panning, and hover tooltips.

  - Ease of Use:
    - Matplotlib requires more code and customization for complex plots, whereas Plotly is more user-friendly for interactive visualizations.

  - Output Type:
    - Matplotlib is mainly used for static images suitable for reports and publications. Plotly is ideal for web-based and dashboard applications.

  - Performance with Large Data:
    - Plotly handles large datasets better in interactive environments compared to Matplotlib.

10. What is the significance of hierarchical indexing in Pandas?
  - Hierarchical indexing allows multiple index levels, making it easier to work with complex, high-dimensional data.

11. What is the role of Seaborn's pairplot() function?
  - pairplot() creates pairwise plots to show relationships between multiple variables in a dataset.

12. What is the purpose of the describe() function in Pandas?
  - describe() provides summary statistics like count, mean, min, max, and quartiles.

13. Why is handling missing data important in Pandas?
  - Missing data can cause incorrect results. Handling it ensures accurate analysis and reliable models.

14. What are the benefits of using Plotly for data visualization?
  - Interactive charts
  - Web-friendly
  - Supports dashboards
  - Easy to explore large datasets

15. How does NumPy handle multidimensional arrays?
  - NumPy handles multidimensional data using ndarrays (N-dimensional arrays). These arrays can store data in one, two, or multiple dimensions efficiently.
    - Data is stored in contiguous memory blocks, which allows fast access and computation.
    - NumPy supports indexing and slicing across multiple dimensions.
    - It provides built-in functions for reshaping, transposing, and performing mathematical operations on multidimensional arrays.

16. What is the role of Bokeh in data visualization?
  - Bokeh is used to create interactive, web-based visualizations for large and streaming datasets.

17. Explain the difference between apply() and map() in Pandas.
  - map()
    - Used only with a single column (Series).
    - Applies a function to each value one by one.
    - Simple and easy to use.
    - Example: changing values in one column.
  - apply()
    - Used with rows or columns.
    - Can work on multiple columns.
    - Useful for slightly complex calculations.
    - Example: calculating a new value using more than one column.

18. What are some advanced features of NumPy?
  - Broadcasting
  - Linear algebra functions
  - Random number generation
  - Memory optimization
  - FFT (Fast Fourier Transform)

19. How does Pandas simplify time series analysis?
  - Pandas provides date indexing, resampling, rolling averages, and time-based filtering.

20. What is the role of a pivot table in Pandas?
  - Pivot tables summarize data by grouping and aggregating values for better insights.

21. Why is NumPy's array slicing faster than Python's list slicing?
  - NumPy arrays store data in contiguous memory, while Python lists store references to objects.
  - NumPy operations are implemented in C, making them faster than Python's interpreted loops.
22. What are some common use cases for Seaborn?
  - Seaborn is mainly used for statistical data visualization. Some common use cases are:
    - Showing data distribution (histograms, KDE plots)
    - Comparing categories (bar plots, box plots, violin plots)
    - Finding relationships between variables (scatter plots, pair plots)
    - Visualizing correlations (heatmaps)
    - Regression analysis (regression plots)

```python
# How do you create a 2D NumPy array and calculate the sum of each row?
import numpy as np

arr = np.array([[1, 2, 3,4],
                [4, 5, 6,7]])

row_sum = arr.sum(axis=1)
print(row_sum)
```

```python
# Write a Pandas script to find the mean of a specific column in a DataFrame.
import pandas as pd

df = pd.DataFrame({'Marks': [72, 80, 94]})

mean_value = df['Marks'].mean()
print(mean_value)
```

```python
# Create a scatter plot using Matplotlib.
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y = [10, 25, 25, 30, 45]

plt.scatter(x, y)
plt.xlabel("X values")
plt.ylabel("Y values")
plt.title("Scatter Plot")
plt.show()
```

```python
# How do you calculate the correlation matrix using Seaborn and visualize it with a heatmap.
import seaborn as sb
import pandas as pd
import matplotlib.pyplot as plt

df = pd.DataFrame({
    'A': [1, 2, 3],
    'B': [4, 5, 6],
    'C': [7, 8, 9]
})

corr = df.corr()
sb.heatmap(corr, annot=True)
plt.show()
```

```python
# Generate a bar plot using Plotly.
import plotly.express as px

data = {'Fruits': ['Apple', 'Banana', 'Orange','Cherry','Kiwi'],
        'Quantity': [10, 15, 7, 8, 13]}

fig = px.bar(data, x='Fruits', y='Quantity', title="Bar Plot")
fig.show()
```

```python
# Create a DataFrame and add a new column based on an existing column.
import pandas as pd

df = pd.DataFrame({
    'Salary': [30000, 40000, 50000]
})

df['Bonus'] = df['Salary'] * 0.3
print(df)
```

```python
# Write a program to perform element-wise multiplication of two NumPy arrays.
import numpy as np

a = np.array([1, 2, 3])
b = np.array([4, 5, 6])

result = a * b
print(result)
```

```python
# Create a line plot with multiple lines using Matplotlib.
import matplotlib.pyplot as plt

x = [1, 2, 3, 4]
y1 = [10, 20, 30, 40]
y2 = [15, 25, 35, 45]

plt.plot(x, y1, label='Line 1')
plt.plot(x, y2, label='Line 2')
plt.legend()
plt.show()
```

```python
# Generate a Pandas DataFrame and filter rows where a column value is greater than a threshold.
import pandas as pd

df = pd.DataFrame({
    'Score': [45, 60, 75, 30]
})

filtered_df = df[df['Score'] > 50]
print(filtered_df)
```

```python
# Create a histogram using Seaborn to visualize a distribution.
import seaborn as sns
import matplotlib.pyplot as plt

data = [10, 20, 20, 30, 40, 50]

sns.histplot(data, bins=5)
plt.show()
```

```python
# Perform matrix multiplication using NumPy.
import numpy as np

A = np.array([[1, 2],
              [3, 4]])

B = np.array([[5, 6],
              [7, 8]])

result = np.dot(A, B)
print(result)
```

```python
# Create a 3D scatter plot using Plotly.
import plotly.express as px
import pandas as pd

df = pd.DataFrame({
    'x': [1, 2, 3, 4],
    'y': [10, 20, 30, 40],
    'z': [5, 15, 25, 35]
})

fig = px.scatter_3d(df, x='x', y='y', z='z', title="3D Scatter Plot")
fig.show()
```
